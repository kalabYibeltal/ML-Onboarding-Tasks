```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/content/imdb_top_1000.csv')
```

```python
genre_counts = data['Genre'].value_counts()

plt.figure(figsize=(10, 6))
genre_counts.plot(kind='bar')
plt.ylabel('Genre')
plt.xlabel('Number of Movies')
plt.title('Genre Distribution of Movies')
plt.show()
```

```python
movies_per_year = data.groupby('Released_Year').size()

# print(movies_per_year)

plt.figure(figsize=(10, 6))
plt.scatter(movies_per_year.index, movies_per_year.values)
plt.xlabel('Year')
plt.ylabel('Number of Movies')
plt.title('Number of Movies vs. Year')
plt.show()
```
